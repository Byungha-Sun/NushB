<쓰레드 생성>  
threading.Thread(target=param1, args=param2)  
  
param1: 이 쓰레드가 실행할 코드를 함수로 만들어서 파라메터로 지정
  
param2: param1에서 지정한 함수가 파라메터가 있다면 param2에서 그 인자를 튜플형태로 전달
  
<쓰레드 시작>  
th.start(): 쓰레드를 ready 상태로 만듬  
  
<쓰레드 대기상태 만들기>
time.sleep(대기시간): 쓰레드를 wait(대기) 상태로 만듬  
t시간이 끝나면 다시 ready상태로 가서 줄선다.  
자기 순서 다시 돌아오면 cpu 획득해서 다시 run상태가 됨

```python
import threading, time

def f():
    for i in range(0,26):
        print('th: ', i)
    
def main():
    th = threading.Thread(target=f) # Thread 생성
    th.start()

    s= 'abcdefghijklmnopqrstuvwxyz'
    for i in s:
        print('main: ', i)


main()
```

```python
s=['aaa','bbb','ccc','ddd','eee','fff','ggg','hhh','iii','jjj']
# 2개의 작업이 동시에 실행된다
# 작업1은 키보드 입력 받아서 입력 받은 메시지 출력(10번)
# 작업2는 리스트 s에서 문자열 1초마다 1개씩 읽어서 출력

def f1():
    for i in range(10):
        s=input('메시지 입력: ')
        print(i+1,'번째 메시지: ', s)
        time.sleep(8)
        
def f2():
    for i in s:
        print(i)
        time.sleep(1)
        
def main():
    th1=threading.Thread(target=f1)
    th1.start()
    
    th2=threading.Thread(target=f2)
    th2.start()
    

main()
```

동기화 처리  
  
공유자원: 여러개의 쓰레드가 공유해서 사용하는 데이터나 자원. 동기화처리 필요  
동기화: 배타적 접근 제공, 공유자원에 동시에 접근하는 쓰레드들을 관리한다

```python
import socket
import time

host= '192.168.35.39' #서버주소
port = 3333           #서버 포트번호

#서버소켓오픈() ->클라이언트가 들어올 수 있도록 대문을 열어둠: socket.AF_INET: 주소종류(ip)
#socket.SOCK_STREAM: 통신종류. (UDP, TCT). socket.SOCK_STREAM는 TCP 통신
server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

#여러번 ip.port를 바인드하면 에러나므로 그걸 방지하기 위한 설정.
server_socket.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)

#server socket에 ip와 port 붙여줌(바인드)
server_socket.bind((host,port))

#클라이언트 접속 준비 완료
server_socket.listen()

print('echo server start')

#accept(): 클라이언트 접속 기다리며 대기. 
#클라이언트가 접속하면 서버와 클라이언트가 1:1 통신할 소켓(빨대)와,
#접속한 클라이언트의 주소를 반환
client_soc, addr = server_socket.accept()

print('connected client addr:', addr)

#recv(메시지크기) : 소켓에서 크기만큼 읽음. 
#소켓에 읽을 데이터 없으면 생길때까지 기다림.
data=client_soc.recv(100)
msg=data.decode()#읽은 데이터 디코딩
print('recv msg: ', msg)

time.sleep(5)
server_socket.close()
```
