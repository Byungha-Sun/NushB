```python
##선생님 답안 -1번(가로로 붙히기)
import pandas as pd
data=pd.read_csv('car_accident_age.csv',encoding='euc-kr')
data

a=data[data.columns[1:]].loc[:11]
```

```python
#12줄씩 끊는 코드
# len(data) -> 데이터프레임의 총 행수 

tabs=[]
idx=['1월','2월','3월','4월','5월','6월','7월','8월','9월','10월','11월','12월']
cnt=len(data)//12 # '/'로 나누면 실수로 계산, '//'로 나눠야 정수로 계산
for i in range(0,cnt):
    y=i*12
    title=data['가해자연령층'][y]
    vals=data[data.columns[2:]].loc[y:y+11]
    cols=[[title,title,title,title,title],data.columns[2:]] #2층으로 컬럼 구성
    x=pd.DataFrame(vals.values,columns=cols,index=idx)
    tabs.append(x)
tabs
```

```python
res=tabs[0]
for i in range(1,len(tabs)):
    res=res.join(tabs[i])
res
```

```python
#2
for i in tabs:
    m=i[i.columns[0]].max()
    x=i[i.columns[0]].argmax()
    print(i.columns[0][0],'의 최대 사고건수',m,' / 발생 월: ',idx[x])
```

```python
#3
for i in tabs:
    m=i[i.columns[0]].sum() #사고건수 컬럼의 합계
    x=i[i.columns[1]].sum() #사망자수 컬럼의 합계
    print(i.columns[0][0],'의 전체사고건수',m,' / 총 사망자 ',x)
```

```python
#4
for i in tabs:
    m=i[i.columns[0]].mean() #사고건수 컬럼의 평균
    print(i.columns[0][0],'의 평균사고건수',m,)
```
