```python
import pandas as pd
import numpy as np

data = pd.read_csv('bus.csv', encoding='euc-kr', index_col='노선번호')
data = data.fillna('정보없음')#fillna(채울값): NAN 값을 지정한 값으로 채움
data
```

```python
#data.loc['버스번호']
x = data.loc['9-8'] #시리즈
print('기점:' ,x['기점'])
print('종점:' ,x['종점'])
print('경유지:' ,x['경유지'])
print('첫차시간:' ,x['기점첫차'])
print('막차시간:' ,x['기점막차'])
```
